```python
import pandas

import json

with open("english.json", "r") as file:
    json_data = json.load(file)

df = pandas.DataFrame(list(json_data.items()), columns=["Name", "Translation"])
# df.to_excel("english.xlsx", index=False)
df
```

```python
import pandas

import json

with open("arabic.json", "r", encoding="utf-8") as file:
    json_data = json.load(file)

df = pandas.DataFrame(list(json_data.items()), columns=["Name", "Translation"])
df.to_excel("arabic.xlsx", index=False)
df
```
