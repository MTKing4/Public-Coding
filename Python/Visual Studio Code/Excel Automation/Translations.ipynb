```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time



import pandas as pd
import openpyxl


service = Service(executable_path="chromedriver.exe")
driver = webdriver.Chrome()

# 1. Go to the branches panel page
driver.get("placeholder")

# 2. Set the accessToken in localStorage
access_token = "placeholder"
driver.execute_script(f"window.localStorage.setItem('accessToken', '{access_token}');")

# 3. Refresh the page to apply the token
driver.refresh()



driver.get("placeholder")


time.sleep(5)




df = pd.read_excel('translations.xlsx')

translations = []

for item in df['Translation']:
    if item == "Translation":
        pass
    else:
        translations.append(item)

# print(translations)


for dropID in range(0, 217):
    drop_down_element = (By.ID,"undefined_accordionheader_"+ str(dropID))
    WebDriverWait(driver, 50).until(EC.visibility_of_element_located(drop_down_element))
    drop_down_input = driver.find_element(*drop_down_element)
    drop_down_input.click()
    time.sleep(5)
    print(f"DropID: {dropID}")
    
    textareas = driver.find_elements(By.CSS_SELECTOR, "textarea.p-textarea.p-filled")
    for textarea, message in zip(textareas[:4], translations[:4]):
            driver.execute_script("arguments[0].scrollIntoView({behavior: 'auto', block: 'center'});", textarea)
            textarea.clear()
            textarea.send_keys(message)
            print(f" textbox: {textarea}")
```
